```python
# You can find covid related data under project's data folder 
base_url = "C:\\Users\\arzum\\OneDrive\\Belgeler\\Ada-Lovelace-Academy\\exercises\\data\\covid\\"
infected_dataset_url = base_url + "time_series_covid19_confirmed_global.csv"
recovered_dataset_url = base_url + "time_series_covid19_recovered_global.csv"
deaths_dataset_url = base_url + "time_series_covid19_deaths_global.csv"
```

#### Solution:

```python
import pandas as pd
import numpy as np
```

```python
# CSV dosyalarını pandas DataFrame' e oku
infected_df = pd.read_csv(infected_dataset_url)
recovered_df = pd.read_csv(recovered_dataset_url)
deaths_df = pd.read_csv(deaths_dataset_url)

# DataFrame' lerin ilk 2 satiri goster
print("Infected DataFrame:")
print(infected_df.head(2))

print("\nRecovered DataFrame:")
print(recovered_df.head(2))

print("\nDeaths DataFrame:")
print(deaths_df.head(2))
```

```python
# En son tarihli sütunun adını al
latest_date_column = infected_df.columns[-1]

# 'Country/Region' sütunu üzerinde gruplama yaparak toplam enfekte olan sayıları bulma
total_infected_by_country = infected_df.groupby('Country/Region')[latest_date_column].sum().reset_index()

# Ölüm sayılarını bulma
total_deaths_by_country = deaths_df.groupby('Country/Region')[latest_date_column].sum().reset_index()

# Toplam iyileşen sayıları bulma
total_recovered_by_country = recovered_df.groupby('Country/Region')[latest_date_column].sum().reset_index()

# Her bir ülkedeki toplam enfekte olan sayıları gösterme
print("Total Infected by Country:")
print(total_infected_by_country)

# Toplam ölüm sayılarını gösterme
print("\nTotal Deaths by Country:")
print(total_deaths_by_country)

# Toplam iyileşen sayıları gösterme
print("\nTotal Recovered by Country:")
print(total_recovered_by_country)
```
